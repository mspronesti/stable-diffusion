## Run

```python
from stable_diffusion_pt import pipeline
from PIL import Image

prompt = "A digital illustration of a medieval town, 4k, detailed, trending in \
artstation, fantasy"

prompts = [prompt]

uncond_prompt = "" 
uncond_prompts = [uncond_prompt] if uncond_prompt else None


strength = 0.8

do_cfg = True 
cfg_scale = 7.5  
height = 512  
width = 512  
sampler = "k_lms"  
n_inference_steps = 50  

device = 'cuda'

pipeline.generate(
    prompts=prompts, 
    uncond_prompts=uncond_prompts,
    strength=strength,
    do_cfg=do_cfg, 
    cfg_scale=cfg_scale,
    height=height, 
    width=width, 
    sampler=sampler,
    n_inference_steps=n_inference_steps, 
    device=device, 
    idle_device='cpu'
)[0]
```
